```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Importing the dataset
df = pd.read_csv('train.csv')
```

```python
# Importing the dataset
df_test = pd.read_csv('test.csv')
```

```python
# Correcting some wrong values
df['Electrical'].fillna('SBrkr', inplace=True)
df_test['SaleType'].fillna('WD', inplace=True)
df['MasVnrType'].fillna('None', inplace=True)
df_test['MasVnrType'].fillna('None', inplace=True)
df_test['KitchenQual'].fillna('TA', inplace=True)
df_test['GarageCars'].fillna(0, inplace=True)
df_test['GarageArea'].fillna(0, inplace=True)
df_test['Exterior1st'].fillna('Wd Sdng', inplace=True)
df_test['Exterior2nd'].fillna('Wd Sdng', inplace=True)

# Filling NaNs 
cols_fill_obj = ['PoolQC', 'MiscFeature','Alley','Fence','FireplaceQu','GarageType','GarageCond','GarageYrBlt',
            'GarageFinish', 'GarageQual', 'BsmtExposure','BsmtFinType2','BsmtFinType1','BsmtCond','BsmtQual',
                'LotShape', 'LandContour', 'LotConfig', 'LandSlope']

cols_fill_num = ['LotFrontage', 'MasVnrArea', 'BsmtUnfSF', 'TotalBsmtSF', 'BsmtFinSF2', 'BsmtFinSF1']

df.loc[:,cols_fill_obj] = df.loc[:,cols_fill_obj].fillna('0')
df_test.loc[:,cols_fill_obj] = df_test.loc[:,cols_fill_obj].fillna('0')

df.loc[:,cols_fill_num] = df.loc[:,cols_fill_num].fillna(0)
df_test.loc[:,cols_fill_num] = df_test.loc[:,cols_fill_num].fillna(0)

df.drop(['Utilities'], axis=1, inplace=True)
df_test.drop(['Utilities'], axis=1, inplace=True)
```

```python
# Not categorical
df['MasVnrArea'] = pd.to_numeric(df['MasVnrArea'])
df_test['MasVnrArea'] = pd.to_numeric(df_test['MasVnrArea'])
df['GarageYrBlt'] = pd.to_numeric(df['GarageYrBlt'])
df_test['GarageYrBlt'] = pd.to_numeric(df_test['GarageYrBlt'])
df['LotFrontage'] = pd.to_numeric(df['LotFrontage'])
df_test['LotFrontage'] = pd.to_numeric(df_test['LotFrontage'])

# Getting list of categorical columns
categorical_cols = df.columns.groupby(df.dtypes == 'object')[1]
categorical_cols
```

```python
categorical_cols = categorical_cols.delete(0)
categorical_cols
```

```python
# Taking care of categorical values
from sklearn.preprocessing import LabelEncoder
for i in range(0, len(categorical_cols)):
    labelencoder_X = LabelEncoder()
    df[categorical_cols[i]] = labelencoder_X.fit_transform(df[categorical_cols[i]].astype(str))
    df_test[categorical_cols[i]] = labelencoder_X.transform(df_test[categorical_cols[i]].astype(str))
    del labelencoder_X
```

```python
df.head()
```

```python
# 'LotFrontage' filling empty cells
df_lot = pd.merge(df, df_test, how='outer')
```

```python
# Arranging datasets by existence of 'Deck' value
train_set = df_lot[df_lot['LotFrontage'] != 0]
test_set  = df_lot[df_lot['LotFrontage'] == 0]

# Arranging X-training and X-testing datasets
X_train = train_set[['1stFlrSF', 'TotalBsmtSF', 'TotRmsAbvGrd', 'GrLivArea', 'GarageArea', 'OverallQual',
                     'GarageCars', 'LotShape', 'BsmtUnfSF', 'BedroomAbvGr', 'RoofStyle', 'FullBath', 'PoolQC', 'PoolArea',
                     'MasVnrArea']]
X_test  = test_set[['1stFlrSF', 'TotalBsmtSF', 'TotRmsAbvGrd', 'GrLivArea', 'GarageArea', 'OverallQual',
                     'GarageCars', 'LotShape', 'BsmtUnfSF', 'BedroomAbvGr', 'RoofStyle', 'FullBath', 'PoolQC', 'PoolArea',
                     'MasVnrArea']]

# Arranging y-training datasets
y_train = train_set['LotFrontage']
df_test.shape
```

```python
test_set.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(criterion = 'gini', random_state = 123)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
# Indroduction of predicted data to the main dataset 'df'
y_pred = pd.DataFrame(y_pred)
y_pred = round(y_pred, 0)
y_pred.columns = ['LotFrontage']
```

```python
temp_id = pd.DataFrame(X_test.index)
temp_id.columns = ['temp_id']

y_pred = pd.concat([y_pred, temp_id], axis=1)
y_pred.set_index(['temp_id'], inplace=True)

df_pred = pd.concat([X_test, y_pred], axis=1)
df_pred.head()
```

```python
# Filling the missing cells with deck
df_pred_ind = list(df_pred.index)
for i in range(0, df_test.shape[0]):
    for j in df_pred_ind:
        if (np.isnan(df_test['LotFrontage'][i])) and (df_test['Id'][i] == df_pred['Id'][j]):
            df_test['LotFrontage'][i] = df_pred['LotFrontage'][j]
```

```python
# Data scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X = sc_X.fit_transform(X)
X_test = sc_X.transform(df_test)
```

```python
# Splitting the dataset into the Training set and Test set
X_train = X.copy()
y_train = y.copy()
```

# Building a model

```python
# Gradient Boosting Regression
from sklearn.ensemble import GradientBoostingRegressor
regressor = GradientBoostingRegressor(n_estimators = 100, max_depth = 3, min_samples_split = 2, learning_rate = 0.1)
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_ind = pd.DataFrame(df_test['Id'], index=None)
y_pred = round(pd.DataFrame(y_pred),1)
y_ind = pd.DataFrame(y_ind)
result = pd.concat([y_ind, y_pred], axis=1)
result.columns = ['Id','SalePrice']
result.set_index(['Id'], inplace=True)
result.head(10)
```
